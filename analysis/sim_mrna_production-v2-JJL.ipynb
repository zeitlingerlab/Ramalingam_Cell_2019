```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get all of the simulation rate parameters (rates in per second)
mdrate=0.0042 
iarate=0.19*mdrate
airate=5.1*mdrate
amrate=571*mdrate
frames=10000000
frametime=0.01
```

```python
#now convert everything to frame time units (unit is now 1/frame)
iarate=iarate*frametime
airate=airate*frametime
amrate=amrate*frametime
mdrate=mdrate*frametime
ontime=[1.0/iarate,1.0/airate]
print('on times:')
print(ontime)
print('prod tau:')
print(1.0/amrate)
```

```python
state=False
traj=np.zeros(frames,dtype=float)
states=np.zeros(frames,dtype=float)
for i in range(1,frames):
    if(np.random.exponential(ontime[state])<1.0):
        state=not state
    dm=-mdrate*traj[i-1]
    if(state):
        states[i]=1.0
        dm=dm+amrate
    traj[i]=traj[i-1]+dm
```

```python
xvals=np.arange(frames)*frametime
start=0
stop=10000000
plt.figure(figsize=(20,10))
plt.plot(xvals[start:stop],traj[start:stop])
plt.plot(xvals[start:stop],states[start:stop]*200,alpha=0.2)
plt.ylim(0,600)
plt.xlabel('Time (minutes)')
plt.ylabel('# of mRNA molecules/cell')
# plt.savefig('Mut-timetrace-04202022-2.pdf', format='pdf')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.hist(traj,bins=100, histtype='step')
plt.show()
```

```python
import pandas as pd
from plotnine import *
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
data = pd.DataFrame({'time':xvals,'cuml_mrna':traj})
fig = ggplot(data,aes('cuml_mrna'))+geom_histogram(bins = 100,fill='blue',color='black')+theme_classic()
fig.save('mut-histogram-04222022.pdf')
```
